```python
types = [
    'choose_1',
    'algebra',
    'fill_in_1',
    'open_response',
    'choose_n',
    'rank',
]
```

```python
type_index_by_name = {name:i for i, name in enumerate(types)}

problem_features = np.zeros((len(problems), len(types)))
for i, problem_type in enumerate(problems["problem_type"]):
    type_index = type_index_by_name[problem_type]
    problem_features[i, type_index] = 1
```

```python
from sklearn.linear_model import Ridge


def train_user_model(user_id):
    user_scores = train_scores[train_scores["user_id"] == user_id]
    user_scores = train_scores[train_scores["user_id"] == user_id]
    if len(user_scores) == 0:
        print(f"User {user_id} has no scores in the train_scores DataFrame.")
        return None
    problem_indexes = [
        problem_index_by_id[problem_id] for problem_id in user_scores["problem_id"]
    ]
    train_data = problem_features[problem_indexes]
    train_label = user_scores["correct"]
    model = Ridge(alpha=0.1)
    model.fit(train_data, train_label)
    return model


# build model for each user
user_model_dict = {}
for user_id in train_scores['user_id'].unique():
    user_model_dict[user_id] = train_user_model(user_id)
```

```python
def predict(user_id, problem_id):
    problem_feature = problem_features[problem_index_by_id[problem_id]].reshape((1, -1))
    pred = user_model_dict[user_id].predict(problem_feature)
    return min(max(pred, 0), 1)
```

```python
from sklearn.metrics import mean_squared_error

def eval_rmse(scores: pd.DataFrame) -> float:
    predictions = np.zeros(len(scores))
    for index, row in enumerate(scores.itertuples(index=False)):
        predictions[index] = predict(row[1], row[0])
    rmse = mean_squared_error(scores["correct"], predictions, squared=False)
    return rmse
```
